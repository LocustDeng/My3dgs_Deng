```python
import os
import sys
import math
import numpy as np
import torch
from PIL import Image
from typing import NamedTuple
from plyfile import PlyData, PlyElement
from colmap_loader import read_extrinsics_binary, read_intrinsics_binary, qvec2rotmat

# RT
def getWorld2View2(R, t, translate=np.array([.0, .0, .0]), scale=1.0): # translatescale1.0
    # 4*4RT
    Rt = np.zeros((4, 4))
    Rt[:3, :3] = R.transpose()
    Rt[:3, 3] = t
    Rt[3, 3] = 1.0 # 441
    # Rt
    C2W = np.linalg.inv(Rt)
    # cxcycz=0,0,0
    cam_center = C2W[:3, 3]
    # 
    cam_center = (cam_center + translate) * scale
    # 
    C2W[:3, 3] = cam_center
    # 
    Rt = np.linalg.inv(C2W)
    return np.float32(Rt)


# camera.py
class ImageInfo:
    #### 
    def __init__(self, cid, image_name, image_path, image, R, T, focalX, focalY, width, height):
        self.cid = cid
        self.image_name = image_name
        self.image_path = image_path
        self.image = image
        self.R = R
        self.T = T
        self.focalX = focalX # 
        self.focalY = focalY
        self.width = width
        self.height = height
        self.fovX = self.processFov(self.focalX, self.width)
        self.fovY = self.processFov(self.focalY, self.height)
        self.znear = 0.01 # 
        self.zfar = 100 # 
        ## 
        self.ViewMatrix = self.processViewMatrix(self.R, self.T) # 
        self.ProjMatrix = self.processProjMatrix(self.znear, self.zfar, self.fovX, self.fovY) # 
        # self.ViewProjMatrix = self.ViewMatrix @ self.ProjMatrix # All
        self.ViewProjMatrix = torch.matmul(torch.tensor(self.ViewMatrix, dtype=torch.float32), self.ProjMatrix)
        # 
        self.trans=np.array([0.0, 0.0, 0.0])
        self.scale=1.0
        self.world_view_transform = torch.tensor(getWorld2View2(R, T, trans, scale)).transpose(0, 1).cuda()
        self.camera_center = self.world_view_transform.inverse()[3, :3]
        
    #### 
    def processFov(self, focal, distance):
        return 2*math.atan(distance/(2*focal))
    
    #### ViewMatrix
    def processViewMatrix(self, R, T):
        Rt = np.zeros((4,4))
        Rt[:3, :3] = R.transpose() # R
        Rt[:3, 3] = T
        Rt[3, 3] = 1.0
        return Rt
    
    #### ProjMatrix, 
    def processProjMatrix(self, znear, zfar, fovX, fovY):
        ## left, rightbottomtop
        tanHalfFovX = math.tan((fovX / 2))
        tanHalfFovY = math.tan((fovY / 2))
        right = tanHalfFovX * znear
        left = -right
        top = tanHalfFovY * znear
        bottom = -top
        ## ProjMatrix
        P = torch.zeros(4, 4)
        z_sign = 1.0
        P[0, 0] = (2 * znear) / (right - left)
        P[1, 1] = (2 * znear) / (top - bottom)
        P[2, 2] = (zfar + znear) / (zfar - znear)
        P[2, 3] = (2 * znear * zfar) / (znear - zfar)
        P[3, 2] = z_sign
        ## 
        return P
    
# 
class BasicPointCloud(NamedTuple):
    points : np.array # x,y,z,n*3
    colors : np.array # 
    normals : np.array # 

# colmap
class GSDataLoader:
    #### pathcolmaprender_dirimages
    def __init__(self, path, reading_dir):
        self.path = path
        self.reading_dir = reading_dir
        self.cameras = [] # 
        self.points_cloud = {} # 
        self.readColmap(path, reading_dir) # 
        
    #### ply
    def fetchPly(self, path):
        plydata = PlyData.read(path) # ply
        vertices = plydata['vertex'] # 
        positions = np.vstack([vertices['x'], vertices['y'], vertices['z']]).T #  np.vstack  xyz  (n, 3)  n 
        colors = np.vstack([vertices['red'], vertices['green'], vertices['blue']]).T / 255.0 #   np.vstack  255.0  [0, 1] 
        normals = np.vstack([vertices['nx'], vertices['ny'], vertices['nz']]).T #  np.vstack 
        return BasicPointCloud(points=positions, colors=colors, normals=normals)
    
    #### colmap
    def processColmap(self, cam_extrinsic, cam_intrinsic, images_folder):
        ### 
        for idx, key in enumerate(cam_extrinsic):
            ## 
            extr = cam_extrinsic[key] # cam_extrinsicImage
            intr = cam_intrinsic[extr.camera_id] # 
            width = intr.width
            height = intr.height # 
            cid = intr.id # 
            ## 
            R = np.transpose(qvec2rotmat(extr.qvec)) # R
            T = np.array(extr.tvec) # T
            ## 
            if intr.model == "SIMPLE_PINHOLE":
                focalX = intr.params[0]
                focalY = intr.params[0]
            elif intr.model == "PINHOLE":
                focalX = intr.params[0]
                focalY = intr.params[1]
            else:
                assert False, "Colmap camera model not handled: only undistorted datasets (PINHOLE or SIMPLE_PINHOLE cameras) supported!"
            ## 
            image_path = os.path.join(images_folder, os.path.basename(extr.name))  # os.path.basename(extr.name) extr.name 
            image_name = os.path.basename(image_path).split(".")[0]  #  image_path 
            image = Image.open(image_path)  #  PIL  Image.open  image_path  Image 
            cam_info = ImageInfo(cid=cid, image_name=image_name, image_path=image_path, image=image, R=R, T=T, 
                                 focalX=focalX, focalY=focalY, width=width, height=height)
            ## 
            self.cameras.append(cam_info)
        ### 
        ## ply
        ply_path = os.path.join(self.path, "sparse/0/points3D.ply")
        self.points_cloud = self.fetchPly(ply_path)
        ''''''
            
    #### colmap
    def readColmap(self, path, reading_dir):
        ### colmapbin
        # imagescameras
        cameras_extrinsic_file = os.path.join(path, "sparse/0", "images.bin")
        cameras_intrinsic_file = os.path.join(path, "sparse/0", "cameras.bin")
        cam_extrinsic = read_extrinsics_binary(cameras_extrinsic_file)
        cam_intrinsic = read_intrinsics_binary(cameras_intrinsic_file)

        ### colmap
        self.processColmap(cam_extrinsic, cam_intrinsic, os.path.join(path, reading_dir))

#### 
l_data = GSDataLoader(r"D:\3DGS\gaussian-splatting(note)\data", "images")
print(l_data.cameras[1].ViewProjMatrix)
print(l_data.points_cloud)
```
